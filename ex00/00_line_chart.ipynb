## connect

```python
import pandas as pd
import sqlite3
```

```python
connect = sqlite3.connect('../data/checking-logs.sqlite')
```

## select

```python
pd.io.sql.read_sql("select datetime from pageviews where uid LIKE 'user_%'", con=connect)
```

## create db

```python
df = pd.io.sql.read_sql("select strftime('%Y-%m-%d', datetime) as date, count(*) as views from pageviews where uid LIKE 'user_%' group by date", con=connect)
```

```python
df.plot(fontsize=8, figsize=(15, 8), rot=90, x='date').set_title('Views per day');
```

## close connect

```python
connect.close()
```
