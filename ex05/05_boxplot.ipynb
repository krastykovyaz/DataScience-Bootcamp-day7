```python

import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/ab-test.csv', sep=',')
df
```

```python
df_after = df[df.time == 'after']
df_before = df[df.time == 'before']
```

```python
df_after
```

```python
fig, ax = plt.subplots(1, 2, sharey=True, figsize=(15, 8))
df_before.boxplot(column='diff', by='group', ax=ax[1],
              boxprops=dict(linestyle='-', linewidth=3, color='green'),
              medianprops=dict(linestyle='-', linewidth=2, color='r'),
              whiskerprops=dict(linestyle='-', linewidth=2, color='r'),
              capprops=dict(linestyle='-', linewidth=2, color='r'))
df_after.boxplot(column='diff', by='group', ax=ax[0],
               boxprops=dict(linestyle='-', linewidth=3, color='green'),
               medianprops=dict(linestyle='-', linewidth=2, color='r'),
               whiskerprops=dict(linestyle='-', linewidth=2, color='r'),
               capprops=dict(linestyle='-', linewidth=2, color='r'))
fig.suptitle('Lag before lab deadline. After and before the Newsfeed.',
             fontdict={'fontsize': 15})
ax[1].set_title('after')
ax[0].set_title('before')
plt.show()
```
